```python
n = input("Enter number: ")
if list(n) == sorted(n):
    print("Ascending")
else:
    print("Not ascending")
```

```python
n = input("Enter number: ")
if list(n) == sorted(n, reverse=True):
    print("Descending")
else:
    print("Not descending")
```

```python
n = input("Enter number: ")
if len(n) > 1:
    n = n[-1] + n[1:-1] + n[0]
print("After swap:", n)
```

```python
n = input("Enter number: ")
for digit in set(n):
    print(digit, ":", n.count(digit))
```

```python
n = input("Enter number: ")
even = all(int(d) % 2 == 0 for d in n)
odd = all(int(d) % 2 != 0 for d in n)
if even:
    print("All even")
elif odd:
    print("All odd")
else:
    print("Mixed")
```

```python
n = input("Enter number: ")
even_sum = sum(int(d) for d in n if int(d) % 2 == 0)
odd_sum = sum(int(d) for d in n if int(d) % 2 != 0)
print("Difference:", even_sum - odd_sum)
```

```python
n = input("Enter number: ")
result = n.replace('0', '')
print("Without zeros:", result)
```

```python
n = input("Enter number: ")
sq_sum = sum(int(d)**2 for d in n)
cube_sum = sum(int(d)**3 for d in n)
print("Square sum:", sq_sum)
print("Cube sum:", cube_sum)
```
